```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("surv_variants.csv")
df.head()
```

```python
# 1. Bar Chart: Variant Counts
variant_counts = df['variant'].value_counts().reset_index()
variant_counts.columns = ['variant', 'count']
fig = px.bar(variant_counts, x='variant', y='count', labels={'variant':'Variant', 'count':'Count'}, title='Variant Counts')
fig.show()
```

This bar chart shows how many times each COVID-19 variant appears in the dataset. Each bar represents a different variant, and the height of the bar shows the number of records for that variant. This helps us quickly see which variants are most and least common in the data.



```python
# 2. Pie Chart: Variant Distribution
fig = px.pie(df, names='variant', title='Variant Distribution')
fig.show()
```

The pie chart displays the proportion of each COVID-19 variant in the dataset. Each slice represents a variant, and the size of the slice shows its share of the total records. This makes it easy to compare the relative frequency of each variant visually.

```python
# 3. Line Chart: Total Cases Over Time (by censure_date)
df_sorted = df.sort_values('censure_date')
fig = px.line(df_sorted, x='censure_date', y='total_cases', color='variant', title='Total Cases Over Time by Variant')
fig.show()
```

This line chart shows how the total number of cases changes over time for each variant. The x-axis represents the date (censure_date), and the y-axis shows the total cases. Each line is a different variant. This helps us understand trends, such as which variants increased or decreased over time.



```python
# 4. Scatter Plot: Total Cases vs. Total Deaths
fig = px.scatter(df, x='total_cases', y='total_deaths', color='variant', title='Total Cases vs. Total Deaths by Variant')
fig.show()
```

The scatter plot compares the total number of cases to the total number of deaths for each record, colored by variant. Each point represents a data entry. This graph helps us see if there is any relationship between cases and deaths, and whether some variants are associated with higher deaths.

```python
# 5. Histogram: Total Cases Distribution
fig = px.histogram(df, x='total_cases', nbins=30, title='Total Cases Distribution')
fig.show()
```

This histogram shows how the total cases are distributed across all records. The x-axis shows ranges of total cases, and the y-axis shows how many records fall into each range. This helps us see if most records have low, medium, or high case numbers.

```python
# 6. Box Plot: Mortality Rate by Variant
fig = px.box(df, x='variant', y='mortality_rate', title='Mortality Rate by Variant')
fig.show()
```

The box plot displays the spread of mortality rates for each variant. Each box shows the median, quartiles, and possible outliers for mortality rates within a variant. This helps us compare how deadly each variant is and see which variants have higher or more variable mortality rates.

```python
# 7. Area Chart: Cumulative Total Cases Over Time
df_sorted['cumulative_cases'] = df_sorted['total_cases'].cumsum()
fig = px.area(df_sorted, x='censure_date', y='cumulative_cases', title='Cumulative Total Cases Over Time')
fig.show()
```

This area chart shows the cumulative sum of total cases over time. The x-axis is the date, and the y-axis is the running total of cases. The chart helps us see how quickly cases are adding up and if there are periods of rapid increase.

```python
# 8. Treemap: Total Cases by Country and Variant
fig = px.treemap(df, path=['Country', 'variant'], values='total_cases', title='Total Cases by Country and Variant')
fig.show()
```

The treemap visualizes the total number of cases, grouped by country and then by variant. Each rectangle represents a country-variant combination, and its size shows the number of cases. This helps us identify which countries and variants have the most cases.

```python
# 9. Sunburst Chart: Total Deaths by Country and Variant
fig = px.sunburst(df, path=['Country', 'variant'], values='total_deaths', title='Total Deaths by Country and Variant')
fig.show()
```

The sunburst chart shows the total deaths, first grouped by country and then by variant. The inner circle represents countries, and the outer segments represent variants within those countries. The size of each segment shows the number of deaths, helping us see which country-variant combinations are most deadly.

```python
# 10. Animated Line Chart: Total Cases Over Time by Variant
fig = px.line(df_sorted, x='censure_date', y='total_cases', color='variant', title='Total Cases Over Time by Variant', animation_frame='variant')
fig.show()
```

This animated line chart shows how total cases change over time for each variant, with animation frames for each variant. You can see the trend for one variant at a time, making it easier to focus on individual variant patterns and compare their growth over time.

```python
# 11. Bar Chart: Total Deaths by Country
country_deaths = df.groupby('Country')['total_deaths'].sum().reset_index()
fig = px.bar(country_deaths, x='Country', y='total_deaths', title='Total Deaths by Country')
fig.show()
```

This bar chart shows the total number of deaths in each country. Each bar represents a country, and the height shows the sum of deaths. This helps us compare which countries have been most affected in terms of deaths.

```python
# 12. Box Plot: Growth Rate by Variant
fig = px.box(df, x='variant', y='growth_rate', title='Growth Rate by Variant')
fig.show()
```

This box plot shows the distribution of growth rates for each variant. It helps us see which variants are spreading faster and how much the growth rate varies within each variant.

```python
# 13. Scatter Plot: Duration vs. Total Cases
fig = px.scatter(df, x='duration', y='total_cases', color='variant', title='Duration vs. Total Cases by Variant')
fig.show()
```

This scatter plot compares the duration (in days) of each variant's presence to the total number of cases. It helps us see if variants that lasted longer also caused more cases.

```python
# 14. Histogram: Mortality Rate Distribution
fig = px.histogram(df, x='mortality_rate', nbins=30, title='Mortality Rate Distribution')
fig.show()
```

This histogram shows how mortality rates are distributed in the dataset. It helps us see if most records have low, medium, or high mortality rates.

```python
# Total Deaths by Variant (Bar Chart)
variant_deaths = df.groupby('variant')['total_deaths'].sum().reset_index()
fig = px.bar(variant_deaths, x='variant', y='total_deaths', title='Total Deaths by Variant', labels={'variant': 'Variant', 'total_deaths': 'Total Deaths'})
fig.show()
```

This bar chart shows the total number of deaths for each COVID-19 variant. Each bar represents a variant, and the height of the bar shows the sum of deaths caused by that variant. This helps us easily compare which variants have resulted in more deaths.
